```python
#splitting data into dependent and independent variable
x = data.drop(['class'],axis =1)
y = data['class']
```

## Model Building and Evaluation

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

## 1. Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
naive = GaussianNB()
naive.fit(x_train,y_train)
```

```python
y_pred_nb_train = naive.predict(x_train)
y_pred_nb_test = naive.predict(x_test)
```

```python
print("Train Accuracy",accuracy_score(y_train,y_pred_nb_train))
print("Test Accuracy",accuracy_score(y_test,y_pred_nb_test))
```

```python

print(classification_report(y_test,y_pred_nb_test))
```

## 2. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=1000,max_depth=3,oob_score=True)
rf.fit(x_train,y_train)
```

```python
y_pred_rf_train = rf.predict(x_train)
y_pred_rf_test = rf.predict(x_test)
```

```python
print("Train Accuracy",accuracy_score(y_train,y_pred_rf_train))
print("Test Accuracy",accuracy_score(y_test,y_pred_rf_test))
```

```python
print(classification_report(y_test,y_pred_rf_test))
```

## 3. KNN neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error_rate = []
for i in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    y_pred_kn_test = knn.predict(x_test)
    error_rate.append(np.mean(y_pred_kn_test != y_test))
```

```python
plt.figure(figsize=(12,12))
plt.plot(range(1,30),error_rate,color='red',linestyle='dashed',linewidth=2, markersize=12,marker='o',markerfacecolor='blue')
plt.title("Error Rate vs K-Values")
plt.xlabel("K-Values")
plt.ylabel("Error Rate")
plt.show()
# k = 5 6 11
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
y_pred_knn_train = knn.predict(x_train)
y_pred_knn_test = knn.predict(x_test)
```

```python
print("Train Accuracy",accuracy_score(y_train,y_pred_knn_train))
print("Test Accuracy",accuracy_score(y_test,y_pred_knn_test))
```

```python
print(classification_report(y_test,y_pred_knn_test))
```
