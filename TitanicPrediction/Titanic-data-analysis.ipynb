## Data Preprocessing

```python
# Checking the missing values
# 75% data should be not null. 
all_data.isnull().sum()/len(all_data) * 100
```

```python
# Cabin has 77 % missing data so we drop it
all_data = all_data.drop(['Cabin'],axis=1)
```

```python
all_data.isnull().sum()/len(all_data) * 100
```

```python
all_data.head()
```

```python
all_data.describe()
```

```python
# q3 -q1 = IQR (Inter Quartile Range)
# q3 + 1.5 IQR = +ve outlier
# q1-1.5IQR = -ve outlier
sns.boxplot(y='Age',data=all_data)
```

```python
all_data['Age'].median()
```

```python
all_data['Age'] = all_data['Age'].fillna(all_data['Age'].median())
```

```python
all_data.isnull().sum()/len(all_data) * 100
```

```python
sns.boxplot(y='Fare',data=all_data)
```

```python
all_data['Fare'].median()
```

```python
all_data['Fare'] = all_data['Fare'].fillna(all_data['Fare'].median())
```

```python
all_data.isnull().sum()/len(all_data) * 100
```

```python
# categoriacal columns we go for mode
all_data['Embarked'].value_counts()
```

```python
all_data['Embarked'] = all_data['Embarked'].fillna('S')
```

```python
all_data.isnull().sum()/len(all_data) * 100
```

```python
all_data.drop_duplicates()
```

```python
# 'PassengerId' , 'Name' , 'Ticket' , 'Fare'
all_data = all_data.drop(['PassengerId' , 'Name' , 'Ticket' , 'Fare'],axis =1)
```
