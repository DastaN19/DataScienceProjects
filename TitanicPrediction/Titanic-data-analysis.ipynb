## Model Building and Model Evaluation

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train ,y_test = train_test_split(x,y,test_size=0.2,random_state=101)
```

# 1.Logistic Regression



```python
#!pip install sklearn
from sklearn.linear_model import LogisticRegression
# log y = sigmoid(B0 + B1x1+ b2x2)
logit_model = LogisticRegression()
# train the model
logit_model.fit(x_train,y_train)
```

```python
Y_pred_lr_train = logit_model.predict(x_train)
Y_pred_lr_test = logit_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
# TP - actually survived + predicated survived , TN - actually did not survive and we predicted correctly, FP , FN
```

```python
print('confusion_matrix- Logistic Regression')
confusion_matrix(y_test,Y_pred_lr_test)
```

```python
# accuracy = 89 + 56 / (89 +10 +24 +56) - f1 = 2 * rp/r+p
print('classification_report - Logistic Regression')
print(classification_report(y_test,Y_pred_lr_test))
```

```python
print("Training accuracy - Logistic Regression",accuracy_score(y_train,Y_pred_lr_train))
print("*****************************")
print("testing accuracy - Logistic Regression",accuracy_score(y_test,Y_pred_lr_test))
```

## Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(criterion='gini')
```

```python
dtree.fit(x_train,y_train)
```

```python
Y_pred_dt_train = dtree.predict(x_train)
Y_pred_dt_test = dtree.predict(x_test)
```

```python
print('classification_report -  Decision Tree')
print(classification_report(y_test,Y_pred_dt_test))
```

```python
print("Training accuracy - Decision Tree",accuracy_score(y_train,Y_pred_dt_train))
print("*****************************")
print("testing accuracy - Decision Tree",accuracy_score(y_test,Y_pred_dt_test))
```

```python
from sklearn.tree import plot_tree
plot_tree(dtree, filled=True)
plt.show()
```

```python
## Random Forest 
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(criterion='gini',n_estimators=500)
rf.fit(x_train,y_train)
```

```python
Y_pred_rf_train = rf.predict(x_train)
Y_pred_rf_test = rf.predict(x_test)
```

```python
print('classification_report -  Decision Tree')
print(classification_report(y_test,Y_pred_rf_test))
```

```python
print("Training accuracy - Random Forest",accuracy_score(y_train,Y_pred_rf_train))
print("*****************************")
print("testing accuracy - Random Forest",accuracy_score(y_test,Y_pred_rf_test))
```
