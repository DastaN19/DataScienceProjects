```python
# Filtering only positive and negative reviews
# SQL : 
# SELECT * FROM REVIEW WHERE SCORE != 3 LIMIT 5000 DATAPOINT

filtered_data = pd.read_sql_query("""SELECT * FROM REVIEWS WHERE SCORE != 3 LIMIT 5000""", dataset)

```

```python
def partition(x):
    if x<3:
        return 0
    return 1

actualScore = filtered_data['Score']
PositiveNegative = actualScore.map(partition)
filtered_data['Score'] = PositiveNegative

filtered_data.head(10)
```
